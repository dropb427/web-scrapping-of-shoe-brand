```python

count_of_sizes = df['Count of Sizes']
colors = df['Color_Option'].str.split('/', expand=True)
style_code_product_code = df['Product_Code']

# Create a new DataFrame
df_table2 = pd.DataFrame({
    'Count of Sizes': count_of_sizes,
    'colors 1': colors[0],
    'color 2': colors[1] if len(colors.columns) > 1 else None,
    'color3': colors[2] if len(colors.columns) > 2 else None,
    'color4': colors[3] if len(colors.columns) > 3 else None,
    'color5': colors[4] if len(colors.columns) > 4 else None,
    'Product_Code': style_code_product_code
})


df_table2.to_csv("Table2.csv", index=False)
```

2 Write a query to retrieve the number of different colors available for each category.



```python
conn = sqlite3.connect('shoe.db')

# Define the SQL query
query_colors_per_category = '''
    SELECT Categories, COUNT(DISTINCT Num_colors) AS Num_Different_Colors
    FROM Table1
    GROUP BY Categories;
'''


colors_per_category_df = pd.read_sql_query(query_colors_per_category, conn)

conn.close()

print(colors_per_category_df)
```

3 Write a query to find the most expensive men's shoe.

```python
import sqlite3
import pandas as pd


conn = sqlite3.connect('shoe.db')


query_most_expensive_men_shoe = '''
    SELECT Shoe_Name, Price
    FROM Table1
    WHERE Categories = 'Men'
    ORDER BY Price DESC
    LIMIT 1;
'''

most_expensive_men_shoe_df = pd.read_sql_query(query_most_expensive_men_shoe, conn)


conn.close()

print(most_expensive_men_shoe_df)
```

2  Write a query to retrieve products with high star ratings (e.g., 4 stars or above)

```python

conn = sqlite3.connect('shoe.db')

# Write DataFrame to SQLite table
df_table3.to_sql('Table3', conn, index=False, if_exists='replace')

# Define the SQL query
query = '''
    SELECT
        *
    FROM
        Table3
    WHERE
        Star_Rating > 4.5;
'''

# Execute the query and fetch results into a new DataFrame
result_df = pd.read_sql_query(query, conn)

# Close the database connection
conn.close()

# Display the result DataFrame
print(result_df)

```

3 Write a query that counts the number of reviews for each product.

```python

conn = sqlite3.connect('shoe.db')


df_table3.to_sql('Table3', conn, index=False, if_exists='replace')


query = '''
    SELECT
        ProductCODE,
        COUNT(ReviewID) AS NumberOfReviews
    FROM
        Table3 
    GROUP BY
        ProductCODE;
'''


result_df = pd.read_sql_query(query, conn)


conn.close()


print(result_df)
```
