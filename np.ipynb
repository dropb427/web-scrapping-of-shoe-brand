```python
try:
    for url in shoe_urls:
        # Navigate to the shoe product page
        driver.get(url)

        try:
            # Find the size container on the current page
            size_container = WebDriverWait(driver, 20).until(
                EC.presence_of_element_located((By.CLASS_NAME, 'css-12whm6j'))
            )

            # Find the size elements on the current page
            size_labels = size_container.find_elements(By.CSS_SELECTOR, 'label')
            size_inputs = size_container.find_elements(By.CSS_SELECTOR, 'input')

            # Initialize lists to store size values
            sizes = []
            available_sizes = []

            # Loop through each size label and input
            for label, input_element in zip(size_labels, size_inputs):
                # Extract size value from the label's text content
                size_value = label.text.strip()

                # Check if the size is available (not disabled)
                is_available = not input_element.get_attribute('disabled')

                # Append the size value to the list of sizes
                sizes.append(size_value)

                # If available, append the size to the list of available sizes
                if is_available:
                    available_sizes.append(size_value)

            # Print the results for the current URL
            print("All Sizes:", sizes if sizes else None)
            print("Available Sizes:", available_sizes if available_sizes else None)
            print("Count of Sizes:", len(sizes) if sizes else None)
            print()

        except (NoSuchElementException, TimeoutException) as e:
            # Handle the case where the size container or elements are not found or the wait times out
            print(f"Size information not found for the URL: {url}")
            print(f"Error: {e}")
            print()

except Exception as e:
    print(f"An error occurred: {e}")
```

```python
# Let's now import all the required libraries.
import selenium
import pandas as pd
from selenium import webdriver
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
import time
from selenium.common.exceptions import TimeoutException
import re
from selenium.webdriver.common.by import By
from collections import Counter
```

```python
# let's first connect to the web driver
driver = webdriver.Chrome()
```

```python
url = "https://us.puma.com/us/en"
driver.get(url)
```

```python
# Find the search input 
search_input = driver.find_element('css selector', '[data-test-id="search-flyout-form-input"]')

# Perform a search 
search_input.send_keys('Shoes')
```

```python
search_submit_button = driver.find_element(By.CSS_SELECTOR, '[data-test-id="search-flyout-form-submit"]')
search_submit_button.click()
```

```python
product_names = []
genders = []

# Find all product elements on the page
product_elements = driver.find_elements(By.CSS_SELECTOR, 'h3.w-full.mobile\\:text-sm.mobile\\:pr-0.font-bold.text-base.pr-5.line-clamp-2')

#  for gender identification
gender_keywords = ['Men\'s', 'Women\'s', 'Girls', 'Boys', 'Unisex']


for product_element in product_elements:
    # Extract product name
    product_name = product_element.text
   
    gender = None

    #  gender keywords in the product name
    for keyword in gender_keywords:
        if keyword in product_name:
            gender = keyword
            break

    # Append the extracted product name to the list
    product_names.append(product_name)

    # Append the extracted gender to the list
    genders.append(gender)

# Print the extracted information
print("Shoe_Name:", product_names)
print("Category:", genders)

```

```python
# Initialize an empty list to store the number of colors
num_colors_list = []

# Find all elements with the specified class
num_colors_elements = driver.find_elements(By.CLASS_NAME, "text-puma-black-300")

# Loop through each element to extract the number of colors
for num_colors_element in num_colors_elements:
    # Extract the text content
    num_colors = num_colors_element.text
    
    # Append the extracted number of colors to the list
    num_colors_list.append(num_colors)

# Print or use the extracted information
print("Num_colors:", num_colors_list)
```

```python
# Initialize an empty list to store prices
price_list = []

# Find all price elements on the page
price_elements = driver.find_elements(By.CSS_SELECTOR, '[data-test-id="price"]')

# Loop through each price element
for price_element in price_elements:
    # Extract the text content
    price_text = price_element.text

    # Append the extracted price to the list
    price_list.append(price_text)

# Print the list of prices
print("Price:", price_list)
```

```python
# Find all elements by class name
review_spans = driver.find_elements(By.CLASS_NAME, 'ml-1.pl-px.text-sm')

# Initialize an empty list to store review counts
review_counts = []

# Loop through each element to extract and append review counts
for review_span in review_spans:
    # Extract the text content
    review_count_text = review_span.text

    
    review_count = re.search(r'\((\d+)\)', review_count_text)
    if review_count:
        review_count = review_count.group(1)
        # Append the extracted review count to the list
        review_counts.append(review_count)

# Print or use the extracted review counts
print("Review:", review_counts)
```

```python
# Find all elements by class name
review_spans = driver.find_elements(By.CLASS_NAME, 'ml-1.pl-px.text-sm')

# Initialize an empty list to store review counts
review_counts = []

# Loop through each element to extract and append review counts
for review_span in review_spans:
    # Extract the text content
    review_count_text = review_span.text

    # Use regular expression to find the review count
    review_count_match = re.search(r'\((\d+)\)', review_count_text)

    # Check if a match is found
    if review_count_match:
        # Extract the review count from the match
        review_count = review_count_match.group(1)

        # Append the extracted review count to the list
        review_counts.append(review_count)
    else:
        # If no match is found, append None
        review_counts.append(None)

# Print or use the extracted review counts
print("Review Counts:", review_counts)
```

```python
shoe_urls = []
elements = driver.find_elements(By.CSS_SELECTOR, '[data-test-id="product-list-item-link"]')

for element in elements[:100]:
    url = element.get_attribute("href")
    shoe_urls.append(url)

print(shoe_urls)
```

```python
# Iterate through each shoe URL
for url in shoe_urls:
    try:
        # Navigate to the shoe product page
        driver.get(url)
        
        # Find the size elements on the current page
        size_labels = driver.find_elements(By.CSS_SELECTOR, '#size-picker label')
        size_inputs = driver.find_elements(By.CSS_SELECTOR, '#size-picker input')

        # Initialize lists to store size values
        sizes = []
        available_sizes = []

        # Loop through each size label and input
        for label, input_element in zip(size_labels, size_inputs):
            # Extract size value from the label's text content
            size_value = label.find_element(By.CSS_SELECTOR, '[data-content="size-value"]').text

            # Check if the size is available (not disabled)
            is_available = not input_element.get_attribute('disabled')

            # Append the size value to the list of sizes
            sizes.append(size_value)

            # If available, append the size to the list of available sizes
            if is_available:
                available_sizes.append(size_value)

        # Print the results for the current URL
        print("All Sizes:", sizes or "None")
        print("Available_Sizes:", available_sizes or "None")
        print("Count_of_Sizes:", len(sizes) if sizes else "None")
        print()

    except Exception as e:
        # Handle the case where an error occurs (no size information found)
        print(f"An error occurred for {url}: {str(e)}")
        print("All Sizes: None")
        print("Available_Sizes: None")
        print("Count_of_Sizes: None")
        print()
```

```python
# Iterate through each shoe URL
for url in shoe_urls:
    # Navigate to the shoe product page
    driver.get(url)

    try:
        # Wait for the element to be present
        style_color_element = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, "ul[data-uds-child='text']"))
        )

        # Find all <li> elements within the <ul>
        li_elements = style_color_element.find_elements(By.TAG_NAME, 'li')

        # Initialize variables to store style code and color
        style_code = None
        color = None

        # Iterate through each <li> element
        for li_element in li_elements:
            # Extract text content from the <li> element
            li_text = li_element.text

            if 'Style' in li_text:
                style_code = li_text.replace('Style: ', '')
            elif 'Color' in li_text:
                color = li_text.replace('Color: ', '')

        # Check if style_code or color is None, and print accordingly
        if style_code is not None:
            print("Product_code:", style_code)
        else:
            print("Product_code: None")

        if color is not None:
            print("Color:", color)
        else:
            print("Color: None")

    except TimeoutException:
        print(f"Timeout while waiting for elements on {url}")
```

```python
import re

# Initialize a dictionary to store star ratings for each product
product_ratings = {}

# Iterate through each shoe URL
for url in shoe_urls:
    driver.get(url)

    # Find all elements by data-test-id
    rating_elements = driver.find_elements(By.CSS_SELECTOR, '[data-test-id="rating-under-title"]')

    # Loop through each element to extract and append star ratings
    for rating_element in rating_elements:
        # Extract the text content
        rating_text = rating_element.text

        # Extract the numeric part of the text using regular expression
        rating = re.search(r'\d+\.\d+', rating_text)
        if rating:
            star_rating = float(rating.group())
            product_ratings[url] = star_rating

# Categorize quality based on star rating
def categorize_quality(star_rating):
    try:
        if star_rating >= 4.5:
            return "Excellent"
        elif star_rating >= 3.5:
            return "Good"
        elif star_rating >= 2.5:
            return "Average"
        elif star_rating >= 1.5:
            return "Poor"
        else:
            return "Very Poor"
    except ValueError:
        return "Not Available"

# Initialize a dictionary to store quality labels for each product
product_quality_labels = {}

# Loop through each product and categorize quality
for product_url, star_rating in product_ratings.items():
    quality_label = categorize_quality(star_rating)
    product_quality_labels[product_url] = quality_label

# Categorize comfort level based on star rating
def categorize_comfort(star_rating):
    try:
        if star_rating >= 4.0:
            return "Very Comfortable"
        elif star_rating >= 3.0:
            return "Comfortable"
        else:
            return "Average"
    except ValueError:
        return "Not Available"

# Initialize a dictionary to store comfort labels for each product
product_comfort_labels = {}

# Loop through each product and categorize comfort
for product_url, star_rating in product_ratings.items():
    comfort_label = categorize_comfort(star_rating)
    product_comfort_labels[product_url] = comfort_label

# Print the result for each URL
for url in shoe_urls:
    print(f"URL: {url}")
    print(f"Star Rating: {product_ratings.get(url, 'Not Available')}")
    print(f"Quality Label: {product_quality_labels.get(url, 'Not Available')}")
    print(f"Comfort Label: {product_comfort_labels.get(url, 'Not Available')}\n")
```
